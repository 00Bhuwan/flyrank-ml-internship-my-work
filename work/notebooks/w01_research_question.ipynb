## 2. The question: decision, action, cost of a wrong call

Unit of analysis:  
-> one row = one content page (content_id), aggregated over its available 90-day window of search/engagement signals.  
Question:  
-> Given a page's pre-decision observable signals (impressions, position, CTR, age, freshness, word count), how likely is it to be declining and worth reviewing first?  
Decision it improves:  
-> which pages a content team spends its limited review/rewrite capacity on this cycle — not "should we review everything" but "who goes to the top of a finite queue."  
Action:  
-> a specific page gets assigned to an editor for refresh, expansion, or monitoring, instead of being picked at random or by gut feel.  
Cost of a wrong call:  
-> a false positive burns editor hours on a page that wouldn't have improved; a false negative lets a genuinely recoverable page keep losing visibility uncaught. Because the review queue is capacity-constrained, ranking quality at the top (precision@K) is what matters — not accuracy across all 30,000 pages.
